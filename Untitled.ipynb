```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
customer=pd.read_csv('customer_data.csv')

customer['gender']= np.where((customer['gender'])=='Male',1, 0)
```

```python
customer.head()
```

```python
customer['Partner'] = customer['Partner'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['Dependents'] = customer['Dependents'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['PhoneService'] = customer['PhoneService'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['OnlineSecurity'] = customer['OnlineSecurity'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['TechSupport'] = customer['TechSupport'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingTV'] = customer['StreamingTV'].apply(lambda x: 0 if x=='No' else 1)
customer['OnlineBackup'] = customer['OnlineBackup'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingMovies'] = customer['StreamingMovies'].apply(lambda x: 0 if x=='No' else 1)
customer['DeviceProtection'] = customer['DeviceProtection'].apply(lambda x: 0 if x=='No' else 1)
customer['PaperlessBilling'] = customer['PaperlessBilling'].apply(lambda x: 0 if x=='No' else 1)
customer['Churn'] = customer['Churn'].apply(lambda x: 0 if x=='No' else 1)
customer['MultipleLines'] = customer['MultipleLines'].apply(lambda x: 1 if x=='Yes' else 0)
customer
```

```python

customer = pd.concat([customer.drop('InternetService',axis=1), pd.get_dummies(customer['InternetService'])],axis=1)
customer = pd.concat([customer.drop('Contract',axis=1), pd.get_dummies(customer['Contract'])],axis=1)
customer = pd.concat([customer.drop('PaymentMethod',axis=1), pd.get_dummies(customer['PaymentMethod'])],axis=1)
```

```python
customer
```

```python
customer['SeniorCitizen'].value_counts().plot('bar')
plt.xlabel('population')
plt.ylabel('number of SeniorCitizen discount')
plt.title('SeniorCitizen discount')
plt.show()
```

```python
customer.groupby('tenure')['tenure'].count().plot()
plt.xlabel('tenure years')
plt.ylabel('number of people')
plt.title('tenure years vs people')
plt.show()
```

```python
no_internet=customer.customerID[(customer['PhoneService'] == 1) &(customer['No'] == 1)].count()
no_phone=customer.customerID[(customer['PhoneService'] == 0) &((customer['DSL'] == 1) |(customer['Fiber optic'] == 1))].count()
phone_internet=customer.customerID[(customer['PhoneService'] == 1) &((customer['DSL'] == 1) |(customer['Fiber optic'] == 1))].count()
list_service=[no_internet,no_phone,phone_internet]
s=[1,2,3]
plt.bar(s,list_service,align='center', alpha=0.5)
objects_service=['no_internet','no_phone','phone_internet']
plt.xticks(s, objects_service)
plt.title("type of services")
plt.xlabel("type of services")
plt.ylabel("Number of people")
plt.show()
```

```python
y=customer.customerID[customer['Fiber optic'] == 1].count()
z=customer.customerID[customer['No'] == 1].count()
list_internet=[x,y,z]
objects=["DSL","fiber","No"]
x=[1,2,3]
plt.bar(x,list_internet,align='center', alpha=0.5)
plt.xticks(x, objects)
plt.title("Internet")
plt.xlabel("Internet types")
plt.ylabel("Number of people")
plt.show()
```

```python
a=customer.customerID[customer['Month-to-month'] == 1].count()
b=customer.customerID[customer['One year'] == 1].count()
c=customer.customerID[customer['Two year'] == 1].count()
list_payment=[a,b,c]
objects_payment=["Month-to-month","One year","Two year"]
t=[1,2,3]
plt.bar(t,list_payment,align='center', alpha=0.5)
plt.xticks(t, objects_payment)
plt.title("Payment plan")
plt.xlabel("payment plan types")
plt.ylabel("Number of people")
plt.show()
```

```python
d=customer.customerID[customer['Bank transfer (automatic)'] == 1].count()
e=customer.customerID[customer['Credit card (automatic)'] == 1].count()
f=customer.customerID[customer['Electronic check'] == 1].count()
g=customer.customerID[customer['Mailed check'] == 1].count()
list_paymenttype=[d,e,f,g]
objects_paymenttype=["Bank transfer (automatic)","Credit card (automatic)","TElectronic check",'Mailed check']
t=[1,2,3,4]
plt.bar(t,list_paymenttype,align='center', alpha=10)
plt.xticks(t, objects_paymenttype)
plt.title("Payment types")
plt.xlabel("payment types")
plt.ylabel("Number of people")
plt.show()
```
