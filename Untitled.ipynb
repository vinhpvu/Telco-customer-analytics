```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
customer=pd.read_csv('customer_data.csv')

customer['gender']= np.where((customer['gender'])=='Male',1, 0)
```

```python
customer.head()
```

```python
customer['Partner'] = customer['Partner'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['Dependents'] = customer['Dependents'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['PhoneService'] = customer['PhoneService'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['OnlineSecurity'] = customer['OnlineSecurity'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['TechSupport'] = customer['TechSupport'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingTV'] = customer['StreamingTV'].apply(lambda x: 0 if x=='No' else 1)
customer['OnlineBackup'] = customer['OnlineBackup'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingMovies'] = customer['StreamingMovies'].apply(lambda x: 0 if x=='No' else 1)
customer['DeviceProtection'] = customer['DeviceProtection'].apply(lambda x: 0 if x=='No' else 1)
customer['PaperlessBilling'] = customer['PaperlessBilling'].apply(lambda x: 0 if x=='No' else 1)
customer['Churn'] = customer['Churn'].apply(lambda x: 0 if x=='No' else 1)
customer['MultipleLines'] = customer['MultipleLines'].apply(lambda x: 1 if x=='Yes' else 0)
customer
```

```python

customer = pd.concat([customer.drop('InternetService',axis=1), pd.get_dummies(customer['InternetService'])],axis=1)
customer = pd.concat([customer.drop('Contract',axis=1), pd.get_dummies(customer['Contract'])],axis=1)
customer = pd.concat([customer.drop('PaymentMethod',axis=1), pd.get_dummies(customer['PaymentMethod'])],axis=1)
```

# Comparing all services in churn customer

For all services , the graph shows that high amount of people leaves the company due to phone services than other services.

```python

churn_DSL=churn_data.customerID[churn_data["DSL"]==1]
churn_Fiber=churn_data.customerID[churn_data["Fiber optic"]==1]
churn_Phone=churn_data.customerID[churn_data["PhoneService"]==1]
churn_onlinesercurity=churn_data.customerID[churn_data["OnlineSecurity"]==1]
churn_onlinebackup=churn_data.customerID[churn_data["OnlineBackup"]==1]
churn_deviceprotect=churn_data.customerID[churn_data["DeviceProtection"]==1]
churn_techsupport=churn_data.customerID[churn_data["TechSupport"]==1]
churn_streamingtv=churn_data.customerID[churn_data["StreamingTV"]==1]
churn_streamingmovies=churn_data.customerID[churn_data["StreamingMovies"]==1]

x_customer=len(customer)
x_DSL=(len(churn_DSL)/float(x_customer))*100
x_Fiber=(len(churn_Fiber)/float(x_customer))*100
x_phone=(len(churn_Phone)/float(x_customer))*100

x_onlinesercurity=(len(churn_onlinesercurity)/float(x_customer))*100
x_onlinebackup=(len(churn_onlinebackup)/float(x_customer))*100
x_deviceprotect=(len(churn_deviceprotect)/float(x_customer))*100
x_techsupport=(len(churn_techsupport)/float(x_customer))*100
x_streamingtv=(len(churn_streamingtv)/float(x_customer))*100
x_streamingmovies=(len(churn_streamingmovies)/float(x_customer))*100



objects = ('DSL', 'Fiber', 'Phone','Online Security', 'Oline Backup', 'Device Protection','Tech Support','Streaming Tv','Streaming Movies')
y_pos = np.arange(len(objects))
performance = [x_DSL,x_Fiber,x_phone,x_onlinesercurity,x_onlinebackup,x_deviceprotect,x_techsupport,x_streamingtv,x_streamingmovies]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)

plt.yticks(size = 7)
plt.xticks(y_pos, objects,size = 6)
plt.ylabel('Percent of people Churn',size = 7)
plt.title('services',size = 7)
plt.show()
```
