```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
customer=pd.read_csv('customer_data.csv')

customer['gender']= np.where((customer['gender'])=='Male',1, 0)
```

```python
customer.head()
```

```python
customer['Partner'] = customer['Partner'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['Dependents'] = customer['Dependents'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['PhoneService'] = customer['PhoneService'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['OnlineSecurity'] = customer['OnlineSecurity'].apply(lambda x: 0 if x=='No' else 1)
```

```python
customer['TechSupport'] = customer['TechSupport'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingTV'] = customer['StreamingTV'].apply(lambda x: 0 if x=='No' else 1)
customer['OnlineBackup'] = customer['OnlineBackup'].apply(lambda x: 0 if x=='No' else 1)
customer['StreamingMovies'] = customer['StreamingMovies'].apply(lambda x: 0 if x=='No' else 1)
customer['DeviceProtection'] = customer['DeviceProtection'].apply(lambda x: 0 if x=='No' else 1)
customer['PaperlessBilling'] = customer['PaperlessBilling'].apply(lambda x: 0 if x=='No' else 1)
customer['Churn'] = customer['Churn'].apply(lambda x: 0 if x=='No' else 1)
customer['MultipleLines'] = customer['MultipleLines'].apply(lambda x: 1 if x=='Yes' else 0)
customer
```

```python

customer = pd.concat([customer.drop('InternetService',axis=1), pd.get_dummies(customer['InternetService'])],axis=1)
customer = pd.concat([customer.drop('Contract',axis=1), pd.get_dummies(customer['Contract'])],axis=1)
customer = pd.concat([customer.drop('PaymentMethod',axis=1), pd.get_dummies(customer['PaymentMethod'])],axis=1)
```

```python
customer.info()
```

```python
churn_data=customer[customer['Churn']==1]
```

# The percent of churn customers in total customers

This graph shows the result of report in which people were churn or not churn. Form the graph, it is clear that the majority of customers are not churn with just 43 percent different to churn customers. Therefore, this company is still doing a good job to keep customers happy with its services. 

```python
churncustomer = len(customer[customer['Churn']==1])
notchurncustomer = len(customer[customer['Churn']==0])
customer_len=len(customer)
churn_customer=churncustomer/float(customer_len)*100
not_churn_customer=notchurncustomer/float(customer_len)*100
x=[1,2]
y=[churn_customer,not_churn_customer]
plt.bar(x, y, color="blue",align = 'center')
labels = ['Churn Customer', 'Not Churn Customer']
plt.xticks(x,labels, color='red')

plt.xlabel('Churn Types')
plt.ylabel('Percent of Total Customers')
plt.title('Churn Graph')
plt.show()
customer.Churn.value_counts()
```

# The percent of Churn customers use phone and internet services.

```python
def percent_column(x):
    percent = len(x)/float (len(churn_data)) * 100
    return percent;
```

The bar chart shows the percent of churn customers in three distinct service groups of phone and DSL, phone and Fiber, and phone only. The percent of customers who use phone and fiber are majority with 69% of total churn customers. The percent of customers, who use phone and DSL, are the second place in chart with 15% of total churn customers. Finally, the percent of customers use phone only are 7 % of total customers churn. Overall, it is clearly that the customers use phone and fiber are likely churn than other service groups. 

```python
churn_phone_DSL=churn_data.customerID[(churn_data["DSL"]==1)&(churn_data["PhoneService"]==1)]
churn_phone_Fiber=churn_data.customerID[(churn_data["Fiber optic"]==1)&(churn_data["PhoneService"]==1)]
churn_phone=churn_data.customerID[(churn_data["PhoneService"]==1)&(churn_data["Fiber optic"]==0)&(churn_data["DSL"]==0)]
                                                                                                 
percent_churn_phone_DSL= percent_column(churn_phone_DSL)
percent_churn_phone_Fiber=percent_column(churn_phone_Fiber)
percent_churn_phone=percent_column(churn_phone)

axis=[1,2,3]
plt.figure(figsize=(10,5))
plt.bar(axis, [percent_churn_phone,percent_churn_phone_DSL,percent_churn_phone_Fiber],color="blue",align = 'center')

labels = ['Phone only','Phone and DSL','Phone and Fiber']

plt.xticks(axis,labels, color='red')

plt.xlabel('Service Types')
plt.ylabel('Percent of Churn Customers')
plt.title('Churn Internet and Phone Services')
plt.show()
```

# The Percent of Churn customers between the Fiber with other services and  the DSL with other services.

The bar chart shows the percent of churn customers in two categories, fiber with other services and DSL with other services. Other services are online security, online backup, device protection, tech support, streaming Tv and streaming movies. Overall, the customers are less likely to churn when they have internet and other services. The percent of fiber with other services is 0.69% and the percent of DSL with other services is 0.1%.

```python
churn_phone_DSL_other=churn_data.customerID[(churn_data["DSL"]==1)&(churn_data["OnlineSecurity"]==1)&(churn_data["OnlineBackup"]==1)
                                      &(churn_data["DeviceProtection"]==1)&(churn_data["TechSupport"]==1)&(churn_data["StreamingTV"]==1)
                                     &(churn_data["StreamingMovies"]==1)]
churn_phone_Fiber_other=churn_data.customerID[(churn_data["Fiber optic"]==1)&(churn_data["OnlineSecurity"]==1)&(churn_data["OnlineBackup"]==1)
                                      &(churn_data["DeviceProtection"]==1)&(churn_data["TechSupport"]==1)&(churn_data["StreamingTV"]==1)
                                     &(churn_data["StreamingMovies"]==1)]

percent_churn_phone_DSL_other= percent_column(churn_phone_DSL_other)
percent_churn_phone_Fiber_other=percent_column(churn_phone_Fiber_other)
axis1=[1,2]
 
plt.bar(axis1, [percent_churn_phone_Fiber_other,percent_churn_phone_DSL_other],color="blue",align = 'center')

labels = ['Fiber with other services' ,'DSL with other services']

plt.xticks(axis1,labels, color='red')

plt.xlabel('Service Types')
plt.ylabel('Percent of Churn Customers')
plt.title('Churn internet and other Services')
plt.show()
print (percent_churn_phone_DSL_other,percent_churn_phone_Fiber_other)
```

# The percent of Churn customer in payment types 

The bar chart illustrates the percent of churn customers in four different payment types, Credit card (automatic),Bank transfer (automatic),Mailed check,Electronic check. The Electronic check  is the most popular payment type with 57% churn customers. The second highest payment type is Mailed check, 16% churn customers. The Credit card (automatic) and Bank transfer (automatic) are mostly the smae with 12% and 13% churn customers.

```python
churn_bank= churn_data[(churn_data["Bank transfer (automatic)"]==1)]
churn_credit= churn_data[(churn_data["Credit card (automatic)"]==1)]
churn_eleccheck= churn_data[(churn_data["Electronic check"]==1)]
churn_mail= churn_data[(churn_data["Mailed check"]==1)]
percent_bank=percent_column(churn_bank)
percent_credit=percent_column(churn_credit)
percent_eleccheck=percent_column(churn_eleccheck)
percent_mail=percent_column(churn_mail)
axisL=[1,2,3,4]
 
plt.bar(axisL, [percent_credit,percent_bank,percent_mail,percent_eleccheck],color="blue",align = 'center')

labels = ['Credit card (automatic)','Bank transfer (automatic)','Mailed check','Electronic check']

plt.xticks(axisL,labels, color='red')

plt.xlabel('Payment Types')
plt.ylabel('Percent of Churn Customers')
plt.title('Churn payment types')
plt.show()
print(percent_credit,percent_bank,percent_mail,percent_eleccheck)
```

# Sernior Discount by Tenure (month)

The histogram chart shows the number of churn customers have senior discounts by tenure (month). The distribution of tenure is skewed right. We have a concentration of data among the early months and a long tail to the right. The vast majority of the customer churn are first five months, with very few customers are churn later in tenure. The first month has the higher number of churn customer, 74 customers compare to other months. The data range from about 1 to about 72, so the approximate range equals 71. There seem to be two probable outliers around 64 to 66.

```python
seniorchurn_data= churn_data[churn_data["SeniorCitizen"]==1]
x_senior = seniorchurn_data['tenure']
plt.figure(figsize=(15,10))
plt.hist([x_senior], bins=72,
         color=["Blue"]) 
plt.xlabel('Sernior Discount by Tenure(month)')
plt.ylabel('Number of Churn Customers')
plt.title('Churn Sernior Discount by Tenure')
plt.show()
```
